###Create External Location

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS datacourselnddl_demo
  URL 'abfss://demo@datacourselnddl.dfs.core.windows.net/'
  WITH (STORAGE CREDENTIAL databricks_course_sc)
  COMMENT 'External location for studing data engineer course';
```
