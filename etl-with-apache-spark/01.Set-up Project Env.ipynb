##2. Create the catalog - gizmobox

```python
%sql
CREATE CATALOG IF NOT EXISTS GIZMOBOX
  MANAGED LOCATION 'abfss://gizmobox@datacourselnddl.dfs.core.windows.net/'
  COMMENT 'Este Ã© o catalogo para o projeto gizmobox Data Lakehouse';
```

##3. Create Schemas
1. Landing
2. Bronze
3. Silver
4.Gold

```python
%sql
CREATE SCHEMA IF NOT EXISTS GIZMOBOX.LANDING
  MANAGED LOCATION 'abfss://gizmobox@datacourselnddl.dfs.core.windows.net/landing'
  COMMENT 'Schema para dados de entrada';
  CREATE SCHEMA IF NOT EXISTS GIZMOBOX.BRONZE
  MANAGED LOCATION 'abfss://gizmobox@datacourselnddl.dfs.core.windows.net/bronze'
  COMMENT 'Schema para dados da bronze';
  CREATE SCHEMA IF NOT EXISTS GIZMOBOX.SILVER
  MANAGED LOCATION 'abfss://gizmobox@datacourselnddl.dfs.core.windows.net/silver'
  COMMENT 'Schema para dados da silver';
  CREATE SCHEMA IF NOT EXISTS GIZMOBOX.GOLD
  MANAGED LOCATION 'abfss://gizmobox@datacourselnddl.dfs.core.windows.net/gold'
  COMMENT 'Schema para dados da gold';
```

##4. Create Volume

```python
%sql
USE CATALOG gizmobox;
USE SCHEMA LANDING;

CREATE EXTERNAL VOLUME IF NOT EXISTS operational_data 
    LOCATION 'abfss://gizmobox@datacourselnddl.dfs.core.windows.net/landing/operation_data/'
    COMMENT 'Operational Data Volume';

CREATE EXTERNAL VOLUME IF NOT EXISTS external_data
    LOCATION 'abfss://gizmobox@datacourselnddl.dfs.core.windows.net/landing/external_data/'
    COMMENT 'External Data Volume';
```
